## 6. Assign

```python
students_score = {
    'Name' : ['Paul', 'Aaron', 'Krista', 'Veronica', 'Paxton', 'Madison', 'Aurora'],
    'Score': [98, 89, 99, 87, 90, 83, 82],
}

df_students = pd.DataFrame(students_score)

df_students
```

```python
grade_series = pd.Series(['AA', 'AB', 'AA', 'AC', 'AB', 'BA', 'BB'])
```

```python
state_series = pd.Series(['TX', 'NY', 'TN', 'MA', 'TN', 'MO', 'OH'])
```

```python
df_students_2 = df_students.assign(Grade=grade_series, State=state_series)
```

```python
df_students_2
```

```python
df_students.assign(Score1=lambda x: x.Score + 1)
```

```python
df_students.assign(Scoremin5=df_students['Score'] - 5)
```
