## 4. Calculate state-wise :

### 4.1. Total number of population.

```python
df_census['Population'].groupby(df_census['State_name']).sum().sort_values(ascending=False)
```

```python
df_census.pivot_table(index='State_name',
                      values='Population',
                      aggfunc='sum').sort_values(by='Population', ascending=False)
```

### 4.2. Total no. of the population with different religions.

```python
df_census.columns
```

```python
df_census.groupby('State_name')[['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains']].sum().sort_values(by='Christians', ascending=False)
```
