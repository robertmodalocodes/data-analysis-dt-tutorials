## 11. Which individual country has the Highest No. of TV Shows ?

```python
df_netflix.Country.sample(5)
```

```python
df_netflix.loc[1806, 'Country'].split(', ')
```

```python
for country in df_netflix['Country']:
    print(country, type(country))
```

```python
len(df_netflix[df_netflix['Country'].isnull()])
```

```python
df_netflix['Country'] = df_netflix['Country'].fillna('Unknown')
```

```python
df_netflix[df_netflix['Country'] == 'Unknown'].sample(5)
```

```python
country_list = []

for country in df_netflix['Country']:
    if len(list(country)) == 1:
        country_list.append(country)
    else:
        for _ in country.split(', '):
            country_list.append(_)
    
```

```python
len(country_list)
```

```python
country_list[:10]
```

```python
pd.Series(country_list).value_counts()

```

```python
countries_dict = dict(pd.Series(country_list).value_counts())
```

```python
countries_dict
```

```python
countries_df = pd.DataFrame({'Country': countries_dict.keys(), \
                             'Count': countries_dict.values()})
```

```python
countries_df.head()
```

```python
countries_df[countries_df['Count'] == countries_df['Count'].max()]
```
