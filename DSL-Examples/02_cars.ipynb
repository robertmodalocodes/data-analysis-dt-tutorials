## 1. Instruction ( For Data Cleaning ) - Find all Null Values in the dataset. If there is any null value in any column, then fill it with the mean of that column.

```python
df_cars_converted.isnull().sum()
```

```python
str_columns = ['Make', 'Model', 'Type', 'Origin', 'DriveTrain', 'MSRP', 'Invoice']

num_columns = ['EngineSize', 'Cylinders', 'Horsepower', 'MPG_City', 'MPG_Highway',
       'Weight', 'Wheelbase', 'Length', 'MSRP_Value', 'Invoice_Value']
```

```python
for column in str_columns:
    df_cars_converted[column].fillna(method='ffill', axis=0, inplace=True)

for column in num_columns:
    df_cars_converted[column].fillna(df_cars_converted[column].mean(), inplace=True)
```

```python
df_cars_converted.isnull().sum()
```
