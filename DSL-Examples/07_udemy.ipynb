## 7. Show all courses of Graphic Design where the price is below 100 ?

```python
df_udemy['price'] = df_udemy['price'].replace(to_replace={'Free' : '0'})
```

```python
df_udemy['price'] = df_udemy['price'].astype(int)
```

```python
# Using where
# Includes NaNs (all records with null values)
df_udemy.loc[df_udemy['subject'] == 'Graphic Design'].where(df_udemy['price'] < 100)
```

```python
# Using multiple conditions
df_udemy[(df_udemy['subject'] == 'Graphic Design') & (df_udemy['price'] < 100)]
```

```python
# Using query
df_udemy.query(expr="subject == 'Graphic Design' & price < 100")
```

## 10. What is the Max. Number of Subscribers for Each Level of courses ?

```python
df_udemy['level'].value_counts()
```

```python
df_udemy['num_subscribers'].groupby(df_udemy['level']) \
                            .agg(['min', 'mean', 'max']) \
                            .sort_values(by='max', ascending=False)
```

```python
df_udemy.groupby('level').max()
```
