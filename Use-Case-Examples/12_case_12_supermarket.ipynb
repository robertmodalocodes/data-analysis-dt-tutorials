## 8. Univariate analysis

```python
# Let's differentiate categorical and numerical columns first
# We're using the length of a single value in each column
categoricals = []
numericals = []

for column in df_super.columns:
    if df_super[column].nunique() > 10:
        numericals.append(column)
    else:
        categoricals.append(column)

```

```python
print(f"Categoricals: {categoricals}")
print("")
print(f"Numericals: {numericals}")
```

### 9.9. Daily sales

```python
df_super.columns
```

```python
df_super['Date'].dt.day_of_week
```

```python
dw_mapping = {
    0: 'Mon',
    1: 'Tue',
    2: 'Wed',
    3: 'Thur',
    4: 'Fri',
    5: 'Sat',
    6: 'Sun'   
}
```

```python
df_super['Date'].dt.day_of_week.map(dw_mapping)
```

```python
df_super['Day of week'] = df_super['Date'].dt.day_of_week.map(dw_mapping)
```

```python
df_super.head(2)
```

```python
df_super['Day of week'].value_counts().plot(kind='bar')
```
