## 14. Check Chest Pain Type

Chest paint type (4 values)
- Value 0: typical angina
- Value 1: atypical angina
- Value 2: non-anginal pain
- Value 3: asymtomatic

```python
sns.countplot(x="cp", data=df_heart)
plt.xticks([0,1,2,3], ["typical angina", "atypical angina", "non-anginal pain", "asymtomatic"])
plt.xticks(rotation=45)
plt.show()
```

## 15. Show The Chest Pain Distribution As Per Target Variable

```python
df_heart.columns
```

```python
sns.countplot(x="cp", hue="target", data=df_heart)
plt.xticks([0,1,2,3], ["typical angina", "atypical angina", "non-anginal pain", "asymtomatic"])
plt.xticks(rotation=45)
plt.legend(labels=["No-Disease", "Disease"])
plt.show()
```

## 16. Show Fasting Blood Sugar Distribution According To Target Variable

```python
sns.countplot(x="fbs", hue="target", data=df_heart)
plt.xlabel("Fasting Blood Sugar (FBS)")
plt.xticks([0,1], ["> 120 mg/dl", "> 120 mg/dl"])
plt.xticks(rotation=0)
plt.legend(labels=["No-Disease", "Disease"])
plt.show()
```

## 20. Plot Continuous Variables

```python
df_heart.columns
```

```python
categorical_val = []
continous_val = []

for column in df_heart.columns:
    if df_heart[column].nunique() <= 10:
        categorical_val.append(column)
    else:
        continous_val.append(column)
```

```python
categorical_val
```

```python
continous_val
```

```python
df_heart.hist(continous_val, figsize=(15, 10))
plt.tight_layout()
```
