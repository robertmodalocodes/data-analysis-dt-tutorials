### 22.2. Finding the unique values of genre

```python
df_imdb.columns
```

```python
# since we have multiple genres for one movie, we perform the analysis in three steps
# step 1:
# split the genre values by comma and store each genre in a list
genre_list_1 = []

for genre_set in df_imdb["Genre"]:
    genre_list_1.append(genre_set.split(','))
```

```python
genre_list_1
```

```python
# step 2: convert the list into one dimensional list
genre_list_1d = []

for genre_set in genre_list_1:
    for genre in genre_set:
        genre_list_1d.append(genre)
```

```python
genre_list_1d
```

```python
# step 3: find the unique values in genre list
genre_unique_list = []

for genre in genre_list_1d:
    if genre not in genre_unique_list:
        genre_unique_list.append(genre)
```

```python
genre_unique_list
```

```python
len(genre_unique_list)
```
