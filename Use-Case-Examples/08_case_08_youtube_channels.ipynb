## 6. Data Cleaning  (Replace '--'  to NaN)

```python
df_youtube.head(20)
```

```python
df_youtube = df_youtube.replace('--', np.nan, regex=True)
```

```python
df_youtube.head(20)
```

## 7. Check Null Values In The Dataset

```python
df_youtube.isnull().sum()
```

```python
# percentage of missing values
percent_missing = df_youtube.isnull().sum() / len(df_youtube) * 100
```

```python
percent_missing
```

```python
sns.heatmap(df_youtube.isnull())
```

```python
# dropping the missing values due to having low percentage (minority)
# axis=0 means we're dropping missing values in rows
# inplace=True means we're overwriting the original dataset with newly modified dataset
df_youtube.dropna(axis=0, inplace=True)
```

```python
sns.heatmap(df_youtube.isnull())
```

## 8. Data Cleaning [ Rank Column ]

```python
df_youtube.tail()
```

```python
df_youtube.dtypes
```

```python
df_youtube['Rank'] = df_youtube['Rank'].str[0:-2]
```

```python
df_youtube.head()
```

```python
df_youtube.tail()
```

```python
df_youtube['Rank'] = df_youtube['Rank'].str.replace(',', '')
```

```python
df_youtube.tail()
```

```python
df_youtube.info()
```

```python
# convert the 'rank' data type to numeric data type
df_youtube['Rank'] = df_youtube['Rank'].astype('int')
```

```python
df_youtube.dtypes
```

## 9. Data Cleaning [ Video Uploads & Subscribers ]

```python
df_youtube.head()
```

```python
df_youtube.dtypes
```

```python
df_youtube['Video Uploads'] = df_youtube['Video Uploads'].astype('int')

df_youtube['Subscribers'] = df_youtube['Subscribers'].astype('int')
```

```python
df_youtube.dtypes
```

## 10. Data Cleaning [ Grade Column ]

```python
df_youtube.head()
```

```python
df_youtube['Grade'].unique()
```

```python
df_youtube['Grade'] = df_youtube['Grade'].map({'A++ ': 5, 'A+ ': 4, 'A ': 3, 'A- ': 2, 'B+ ': 1})
```

```python
df_youtube.head()
```

```python
df_youtube.dtypes
```
